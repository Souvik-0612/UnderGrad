# 1. Particle in a box
Potential form:
$$\psi(x) = \begin{cases} 0 & \text{if $|x|<a$}\\
\infty & \text{else} \end{cases}$$
Time independent Schrodinger equation:
$$\frac{d^2\psi}{dx^2} = -\frac{2mE}{\hbar^2}\psi$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h = 1 #h_bar actually
m = 1 
```

```python
a = 1
V = 0

N = 1000
x, dx = np.linspace(0, a, N, retstep = True)
psi = np.zeros(N); psi[0] = 0
DpsiDt = np.zeros(N); DpsiDt[0] = 1 #It's kind of random
En = [4.93, 19.76,44.63, 79.74, 125.4]
for E in En:
#E Eigen energy
    for i in range(N-1):
        psi[i+1] = psi[i] + dx*DpsiDt[i]
        DpsiDt[i+1] = DpsiDt[i] + dx*(-2*m*E/h**2)*psi[i]
    plt.plot(x, psi)
    plt.show()
```

```python
# Next task: Find the all eigen energies.
E = 80
psi = np.zeros(N); psi[0] = 0; psi[-1] = 1
while abs(psi[-1]) > 0.001:
    E += .01
    for i in range(N-1):
        psi[i+1] = psi[i] + dx*DpsiDt[i]
        DpsiDt[i+1] = DpsiDt[i] + dx*(-2*m*E/h**2)*psi[i]
print(E)
    
```
