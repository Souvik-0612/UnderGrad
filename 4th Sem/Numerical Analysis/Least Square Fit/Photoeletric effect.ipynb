## Strange outcomes from the experiment
* Stopping potential($V_0$) remains same as long as frequency $nu$ is constant. It does not depend on intesity of the light, where classical physics says it indeed depends on intensity.
* There is a minimum or *threshold* frequency $\nu_0$ of the radiation below which no emission of eletron takes place.
* Eletron emission takes place immediately the light shines on the surface, with no detectable time.

Our Hero, Einstein, in 1905, said that (from the Plank's posulate) **quantum** properties are inherent in the nature of eletromagnetic radiation itself, so that light consist of quanta called **photons**, each photon having an energy $E = h \nu$. The photons are so localised that whole quatum energy can be ansorbed by a single atom of the cathod and it's entire energy $h \nu$ is used to eject the an eletron from the anode.But as eletron are so bound in the matalic surface it need a certain amount of energy to espace. This minimum amount of energy is called **work function** of the metal,$W_0$ this equals to $h \nu_0$, where $\nu_0$ is thershold frequncy. It follows that the maximum kinetic energy of a photoelectron is given in tenns of the frequency v by the linear relation, 
\begin{equation} \label{1} \tag{1}
\boxed{h \nu - W_0 = eV_0} 
\end{equation}

After solving more \ref{1} we will end up with 
\begin{equation} \label{2} \tag{2}
\boxed{V_0 = \frac{h}{e} \nu - \frac{h}{e} \nu_0 } 
\end{equation}

So the slope($\tan{\theta}$) in the graph $V_0$ vs $\nu$ will be $\frac{h}{e}$. So
\begin{equation} \label{3} \tag{3}
\boxed{h = e\tan{\theta}} 
\end{equation}

```python
lmbda = np.array([460, 500, 540, 570, 635])
StPoten = -np.array([-1.09, -0.88, -0.74, -0.54, -0.33])

freq = 3*10**8/(lmbda*10**-9)

# Gaussian elemination
Aug = np.array([[sum(freq**2), sum(freq), sum(freq*StPoten)],
[sum(freq), len(freq), sum(StPoten)]], dtype=float)

Aug[1] = Aug[1] - Aug[1][0]*Aug[0]/Aug[0][0] #Elemination

#Back Substitution
c = Aug[1][-1]/Aug[1][-2]
m = (Aug[0][-1] - c*Aug[0][-2])/Aug[0][0]

#For fitting the graph
x = np.linspace(freq[0], 3e14, 100)
y = m*x + c

h = m*1.6*10**-19
print(f"Plank's Constant is {h} Js")

#Threshold frequency
for i in range(len(y)):
    if y[i]<0:
        nu_o = x[i]
        print("Threshold frequency",nu_o, "Hz")
        break

print("Workfunction of the metal", nu_o*h, "Joule")

#Plotting
plt.figure(figsize = (14, 10))
plt.title("Photoeletric effect [Set - 1(40 cm)]", size = 25)
plt.plot(freq, StPoten, "ro", label = "Experimental data")
plt.plot(x, y, label = "Fitting curve")
plt.legend(loc = 'best')
plt.xlabel(r"$\nu $", size = 20)
plt.ylabel(r"$V_0[Volts]$", size = 20)
plt.show()
```
