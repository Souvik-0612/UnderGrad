```python
import numpy as np
import matplotlib.pyplot as plt
```

# Simulation Weldon's dice experiment :

The experiment consit of throwing  12 dices (equivalently you can throw single dice   12 times).
 Count 4 and 5 as success, (probability of success 1/3 and failure 2/3 in each throw ).
Repeat the experiment 26306 times and find 
the ensemble average of the probability of 0,1,2...12 success. Compare it with Binomial distribution.

```python
def Binomial_dis(n, k, p):  
    # n Bernouli trails; k success; p = probability of success
    def fac(a):
        if a <= 1:
            return 1
        else: 
            return a*fac(a-1)
        
    def C(m, r): return fac(m)/(fac(r)*fac(m-r))
    
    return C(n, k)*(p**k)*(1-p)**(n-k)
```

```python
def WeldonDiceExp(N, n_dices):   
    #N = Number of expt; n_dices = no of dices
    ss = np.zeros((N, ), dtype = np.uint8) 
    #uint8 for reducing the space complexity

    for i in range(N):
        #Throwing a dice 12 times
        dice = np.random.choice(np.arange(1, 7), size = (n_dices,))

        success = []
        for d in dice:
            if d == 4 or d == 5:
                success.append(d)  #run time = O(N*6)
        ss[i] = len(success)

    yy = []
    for i in range(n_dices+1):
        x = np.count_nonzero(ss == i)
        yy.append(x/N)
        
    return yy
```

### Try: Diffarent number of experiments
Changing the value of N but I have fixed the value of number of dices i.e, n.

```python
N = 10   # no of expt
n = 12   # no. of dices used

xx = np.arange(0, n+1)
ber  = [Binomial_dis(12,i, 1/3 ) for i in range(n+1)]

#Plotiing
plt.figure(figsize = (18, 20))
for i in range(6):
    plt.subplot(3, 2, i+1)
    yy = WeldonDiceExp(N, n)

    plt.bar(xx, yy, alpha = 0.4, label = "Simualtion")
    plt.plot(xx, ber, "-o", label ="Binomial distribution")
    plt.xticks(xx)
    plt.xlabel("Number of success")
    plt.ylabel("Probability")
    plt.legend()
plt.show()
```
