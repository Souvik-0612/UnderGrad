# Varification of central limit theorem using Collatz distribution

```python
#Collatz conjecture
def number_of_step(n):
    count = 0
    while n != 1:
        if n % 2 == 0:
            n = n/2
        else:
            n = 3*n + 1
        count += 1  
    return count
```

```python
x = np.linspace(1, 10, 10)
y = [number_of_step(i) for i in x]
plt.figure(figsize=(16, 10))
plt.bar(x, y)
plt.xlabel('n')
plt.ylabel('Number of iterartions to reach 1')
plt.title('Distribution of the Collatz conjecture')
plt.show()  
```

```python
x = np.linspace(1, 100, 100)
y = [number_of_step(i) for i in x]
m = np.mean(y)
sd = np.std(y)
print(f"Mean of original distribution {m:.2f} and standard daviation {sd:2f}")
s = np.random.choice(y, (50000, 49))
S = np.mean(s, axis=1)
from collections import Counter
c = Counter(S)
x = list(c.keys())
y = list(c.values())

# Mean of S
M = np.mean(S)
SD = np.std(S)
print(f"Mean of sample distribution {M:.2f} and standard daviation {SD:2f}")
xx = np.linspace(min(x), max(x), 100)

# Gaussian distribution with mean M and standard deviation SD
# yy = np.exp(-(xx - M)**2/(2*SD**2))/(np.sqrt(2*np.pi)*SD)    

#Plotting
# plt.plot(xx, max(y)*yy, 'r-')
plt.bar(x, y)
plt.show()
```
