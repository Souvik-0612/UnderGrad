```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import scienceplots
plt.style.use(['science', 'notebook', 'grid'])
```

```python
#Collatz conjecture
def number_of_step(n):
    count = 0
    while n != 1:
        if n % 2 == 0:
            n = n/2
        else:
            n = 3*n + 1
        count += 1  
    return count
```

```python
x = np.linspace(1, 100, 100)
y = [number_of_step(i) for i in x]
plt.bar(x, y)
plt.xlabel('n')
plt.ylabel('Number of steps')
plt.title('Collatz conjecture')
plt.show()  
```
