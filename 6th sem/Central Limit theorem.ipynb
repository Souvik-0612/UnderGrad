## Weighted dices

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import scienceplots
plt.style.use(['science', 'notebook', 'grid'])
```

```python
X = np.linspace(1, 6, 6)
P = np.array([0.3, 0.2, 0.2, 0.1, 0.1, 0.1])

plt.title("The distribution")
plt.bar(X, P, color='blue', alpha=0.5)
plt.xlabel("Face of the dice")
plt.ylabel("Probability")
plt.show()
```

```python
# Mean of the distribution
mean = np.sum(X * P)
# Variance of the distribution
var = np.sum((X - mean)**2 * P)
print(f"Mean: {mean}")
print(f"Variance: {var}")
```

```python
from collections import Counter
size = 10

SS = []
for _ in range(10000):
    S = sum(np.random.choice(X, size, p=P)) ############
    SS.append(S)
# Group elements in SS and count occurrences
counts = Counter(SS)

# Extract the elements and their counts
elements = np.array(list(counts.keys()))
occurrences = np.array(list(counts.values()))
occurrences = occurrences / np.sum(occurrences)

# Plot a bar graph
plt.bar(elements, occurrences)
plt.xlabel("Elements")
plt.ylabel("Occurrences")
plt.title("Occurrences of Elements in SS")
plt.show()
    
```
