```python
#importing the libraries
import numpy as np
import matplotlib.pyplot as plt
```

# Laticce

```python
# we are making 20 by 20 lattice points to verify the Boltzmann's distribution
a = 100
lattice = 2*np.ones((a, a))

iteration = 10000
# plt.figure(figsize=(20, iteration+2))
for _ in range(iteration):
    # #Plotting
    # plt.subplot(10, 4, _+1)
    # plt.imshow(lattice, cmap='hot')  # 'cmap' specifies the colormap
    # plt.colorbar()  # Show color scale
    # plt.xticks(np.arange(0, a, 1))  # Ensure each x-tick is shown
    # plt.yticks(np.arange(0, a, 1))  # Ensure each y-tick is shown
    
    
    p = np.random.randint(a)
    q = np.random.randint(a)
    r = np.random.randint(a)
    s = np.random.randint(a)
    # Choose two lattice point and exchange the number
    var = lattice[r, s]
    lattice[p, q] += var
    lattice[r, s] -= var
    
    
plt.show()
```

```python
# 2D array to 1D array
lattice_flattened = lattice.flatten()
unique_values, counts = np.unique(lattice_flattened, return_counts=True)

# To display the counts of each unique value
# for value, count in zip(unique_values, counts):
#     print(f"Value {value}: {count} times")

# Plotting
plt.figure(figsize=(18, 8))
plt.bar(unique_values, counts)
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('Count of Each Value in the 2D Lattice Array')
plt.xticks(unique_values)  # Ensure each unique value has a tick
plt.show()
```

```python
# Assuming 'lattice' is your 2D array
plt.figure(figsize=(6, 6))
plt.imshow(lattice, cmap='hot')  # 'cmap' specifies the colormap
plt.colorbar()  # Show color scale
plt.title('2D Lattice Visualization')
plt.show()
```

```python
#there is N >> 1 sites in a 1D lattice,
N = 5000
lattice1 = np.ones((N))

itr = 10000
for _ in range(itr):
    a = np.random.randint(N)
    b = np.random.randint(N)
    var = lattice1[a]
    lattice1[a] -= var
    lattice1[b] += var


unique_values, counts = np.unique(lattice1, return_counts=True)

for c in range(len(counts)):
    print(unique_values[c], "--->",counts[c])
```
