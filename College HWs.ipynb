Let say you have a certain number of particle in 3D space. You have to find the distances between each pair.

```python
import numpy as np
```

```python
# Class for particle. It has position in certesian coordinate (x,y,z)
class Particle():
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z
        
# Function to find the distance
def distance(p, q):
    dis = np.sqrt((p.x-q.x)**2 + (p.y-q.y)**2 + (p.z-q.z)**2)
    return dis
```
